# $\omega$ time series with update rule

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import matplotlib
from numpy import random
import copy
import scipy
```

```python
xnum = 20
ynum = 20
N = xnum*ynum

nrec = 1000

dt = 0.1
trange = dt*np.arange(nrec)
```

```python
# load w data

# for the data with self-circling, use the data under the folder "selfcircle"
# for the data without self-circling, use the data under the folder "no_selfcircle"

with open('w_arr.dat','rb') as f:
    warr = np.fromfile(f,np.float64,sep=' ')
warr = warr.reshape((nrec,N))
    
```

```python
matplotlib.rc("font", **{"family":"serif","serif":["Computer Modern Roman"]})
matplotlib.rc("text", usetex=True)
```

```python
thisw = warr[:,10]

%matplotlib inline

fig, ax = plt.subplots(figsize=(10,5),facecolor=(1,1,1))
plt.plot(trange,thisw)
plt.xlabel("t",fontsize=20)
plt.ylabel("$\omega_{0}$",fontsize=20)
#plt.savefig('wupdate_SC_i10_v1.pdf',dpi=100,bbox_inches='tight')
plt.show()
```
